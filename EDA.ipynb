```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("pheme_dataset.csv")
```

```python
df.head(3)
```

```python
df_event_rumor = df[["event", "tweet_id", "label"]].groupby(["event", "label"]).agg(len) \
    .rename(columns={"tweet_id": "total"}).reset_index()
plt.figure(figsize=(10,7))
plt.title("Proportion of Rumor to Non-rumor Tweets by News Event")
ax = sns.barplot(x = "event", y = "total", hue = "label", data = df_event_rumor)
```

```python
plt.figure(figsize=(13,7))
plt.title("Proportion of Source and Response Tweets by News Event")
ax = sns.countplot(x = "event", hue = "is_source", data = df, palette = "husl")
```

```python
fig, axes = plt.subplots(figsize=(12,7))
ax = sns.boxplot(y = "event", 
            x = "retweet_count",
            hue = "label", 
            data = df)

ax.set_title("Thread Length by Event and Rumor")
```
